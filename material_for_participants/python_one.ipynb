```python
odds = [1, 3, 5, 7]                 # use square brackets separate with commas
print('odds are:', odds)

print('first element:', odds[0])    # access elements using indices: first is 0
print('last element:', odds[3])     
print('"-1"th element:', odds[-1])  # last element -1. Counts backwards from the end
print('"-2"th element:', odds[-2])  # second to last element

names = ['Curie', 'Darwing', 'Turing'] # a list of any data type e.g. str
print('names is originally:', names)
names[1] = 'Darwin'                     # correct our typo
print('final value of names:', names)

name = 'Darwin'                 
name[0] = 'd'                # str can not be corrected the same way -> immutable
```

```python
odds.append(11)             # add element at the end
print('odds after adding a value:', odds)

odds.pop(0)   # pop removes an element in a specific index
print('odds after removing the first element:', odds)

odds.reserve()
print('odds after reversing:', odds)
```

```python
odds = [3, 5, 7]
primes = odds
primes.append(2)
print('primes:', primes)
print('odds:', odds)
```

```python
odds = [3, 5, 7]
primes = list(odds)
primes.append(2)
print('primes:', primes)
print('odds:', odds)
```

```python
odds = [1,3,5,7]
print(odds[0])
print(odds[1])
print(odds[2])
print(odds[3])
```

```python
for number in odds:
    print(number)
print("loop finished")
```

```python
odds = [1,3,5,7,9,11,13,15,17]
for number in odds:
    print(number)
```

```python
for odd in odds:
    print(odd)
```

```python
length = 0
names = ['Curie','Darwin','Turing']
for name in names:
    length = length + 1
print('there are', length,'names in the list.')
```

```python
length = 0
names = ['Curie','Darwin','Turing']
for name in names:
    print('name',name,'length before update', length)
    length = length + 1
    print('length after update',length)
print('there are', length,'names in the list.')
```

```python
name = 'rosalind'
print('Name before loop',name)
for name in ['curie','darwin','turing']:
    print('Name inside the loop', name)
print('Name after loop',name)
```

# Visualizing data

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.loadtxt(fname='data/inflammation-01.csv',delimiter=',')
```

```python
plt.imshow(data)
```

```python
plt.plot(np.mean(data,axis=0))
```

```python
plt.plot(np.max(data, axis=0))
```

```python
plt.plot(np.min(data, axis=0))
```

```python
fig = plt.figure(figsize=(10,3))

axes1 = fig.add_subplot(1,3,1)
axes2 = fig.add_subplot(1,3,2)
axes3 = fig.add_subplot(1,3,3)

axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(np.max(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(np.min(data, axis=0))

plt.savefig('inflammation.png')
```

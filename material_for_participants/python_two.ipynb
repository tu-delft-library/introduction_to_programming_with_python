```python
# f(t) = (x - 32) * (5/9)
fahrenheit_val = 99
celsius_val = (fahrenheit_val - 32) * (5/9)
celsius_val
```

```python
def fahrenheit_to_celsius(temperature_f):
    celsius_val = (temperature_f - 32) * (5/9)
    return celsius_val 
    
```

```python
fahrenheit_to_celsius(99)
```

```python
print('freezing point of water:', fahrenheit_to_celsius(32), 'C')
print('boiling point of water:', fahrenheit_to_celsius(212), 'C')
```

```python
def celsius_to_kelvin(temperature_c):
    return temperature_c + 273.15
```

```python
print('freezing point of water:', celsius_to_kelvin(0),'K')
```

```python
def fahrenheit_to_kelvin(temperature_f):
    temperature_c = fahrenheit_to_celsius(temperature_f)
    temperature_k = celsius_to_kelvin(temperature_c)
    return temperature_k
```

```python
print('boiling point pof water in kelvin:',fahrenheit_to_kelvin(212))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
def visualize(filename):
    data = np.loadtxt(fname= filename, delimiter = ',')

    fig = plt.figure(figsize = (10,3))

    axis1 = fig.add_subplot(1,3,1)
    axis2 = fig.add_subplot(1,3,2)
    axis3 = fig.add_subplot(1,3,3)

    axis1.plot(np.mean(data, axis=0))
    axis1.set_ylabel('average')

    axis2.plot(np.max(data, axis=0))
    axis2.set_ylabel('maximum')

    axis3.plot(np.min(data, axis=0))
    axis3.set_ylabel('minimum')

    plt.show()
```

```python
visualize('data/inflammation-02.csv')
```
